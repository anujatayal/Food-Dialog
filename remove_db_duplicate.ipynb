```python
import re

text = "l1itre "
pattern = r'(\d+(\.\d+)?)\s+(.+)'

match = re.match(pattern, text)

if match:
    number = match.group(1)
    rest_of_string = match.group(3)
    print("Number:", number)
    print("Rest of string:", rest_of_string)
else:
    print("No match found.")
```

```python
import spacy

# Load the English language model in spaCy
nlp = spacy.load('en_core_web_sm')

# Input text
text = "The cake was eaten by the hungry dog. The book was read by the student. The student is reading a book."

# Process the text with spaCy
doc = nlp(text)

# Extract passive sentences
passive_sentences = []

for sentence in doc.sents:
    for token in sentence:
        print(token,token.dep_)
        if "pass" in token.dep_:
            print("token")
            passive_sentences.append(sentence.text)

# Print the identified passive sentences
for sentence in passive_sentences:
    print(sentence)
```
